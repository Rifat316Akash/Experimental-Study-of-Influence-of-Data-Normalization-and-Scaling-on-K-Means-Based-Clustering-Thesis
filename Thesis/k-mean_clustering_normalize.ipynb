```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
import math 
```

```python
dataset = pd.read_csv('iris_data.csv')
data = [];

## class level data transfer to ineger values 

abc = dataset['class']
encoder = LabelEncoder()
encoder.fit(abc)
encoded_Y = encoder.transform(abc)
dataset['class'] = encoded_Y

## separate class level from dataset
data1 = dataset.drop('class',axis = 1)
print(data1)

## data scaling 
mn=preprocessing.MinMaxScaler()
data1['sepalLength'] =mn.fit_transform(data1['sepalLength'].values.reshape(-1, 1))
data1['sepalWidth'] =mn.fit_transform(data1['sepalWidth'].values.reshape(-1, 1))
data1['petalLength'] =mn.fit_transform(data1['petalLength'].values.reshape(-1, 1))
data1['petalWidth'] =mn.fit_transform(data1['petalWidth'].values.reshape(-1, 1))
print(data1)

for i in range(150):
    lst = [];
    lst.append(data1['sepalLength'][i]);
    lst.append(data1['sepalWidth'][i]);
    lst.append(data1['petalLength'][i]);
    lst.append(data1['petalWidth'][i]);
    lst.append(dataset['class'][i]);
    data.append(lst);
print(data);
```

```python
c1 = []
c2 = []
c3 = []

c1 = data[one]
print(c1)

c2 = data[two]
print(c2)

c3 = data[three]
print(c3)
```

```python
def dis(source, dst):
    sum = 0
    for i in range(4):
        x = source[i] - dst[i];
        x = x*x    
        sum += x
    return math.sqrt(sum)   
```

```python
def update(c):
    up = []
    
    for i in range(4):
        sum = 0
        for j in range(len(c)):
            sum += c[j][i]
        up.append(sum/len(c))
    return up
```

```python
cl1 = []
cl2 = []
cl3 = []
for i in range(15):
    cl1 = []
    cl2 = []
    cl3 = []
    for j in range(150):
        d1 = dis(c1, data[j])
        d2 = dis(c2, data[j])
        d3 = dis(c3, data[j])
        m = min(d1,d2,d3)
        
        if (m==d1):
            cl1.append(data[j])
        elif (m==d2):
            cl2.append(data[j])
        else:
            cl3.append(data[j])    
    
    c1 = update(cl1)
    c2 = update(cl2)
    c3 = update(cl3)
    
print(c1)
print(c2)
print(c3)
```
