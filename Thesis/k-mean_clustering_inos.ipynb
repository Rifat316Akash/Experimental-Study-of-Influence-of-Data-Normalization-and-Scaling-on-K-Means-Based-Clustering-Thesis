```python
def dis(source, dst):
    sum = 0
    for i in range(7):
        x = source[i] - dst[i];
        x = x*x    
        sum += x
    return math.sqrt(sum)   
```

```python
def update(c):
    up = []
    
    for i in range(7):
        sum = 0
        for j in range(len(c)):
            sum += c[j][i]
        up.append(sum/len(c))
    return up
```

```python
cl1 = []
cl2 = []
for i in range(15):
    cl1 = []
    cl2 = []
    for j in range(351):
        d1 = dis(c1, data[j])
        d2 = dis(c2, data[j])
        m = min(d1,d2)
        if (m==d1):
            cl1.append(data[j])
        elif (m==d2):
            cl2.append(data[j])     
    c1 = update(cl1)
    c2 = update(cl2)
    
print(c1)
print(c2)
```
