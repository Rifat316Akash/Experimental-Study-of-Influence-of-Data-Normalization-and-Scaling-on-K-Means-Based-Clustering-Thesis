```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('F:/Course Study/Thesis(Musharrat Mam)/iris_data.csv')
df
```

```python
print(df['class'].unique())
df.loc[df['class']=='Iris-setosa', 'class'] = 0
df.loc[df['class']=='Iris-versicolor', 'class'] = 2
df.loc[df['class']=='Iris-virginica', 'class'] = 1
print(df['class'].unique())
```

```python
df_1=df.iloc[0:35,]
df_4=df.iloc[35:50,]
df_2=df.iloc[50:85,]
df_5=df.iloc[85:100,]
df_3=df.iloc[100:135,]
df_6=df.iloc[135:150,]
```

```python
df_train=pd.concat([df_1, df_2, df_3])
df_test=pd.concat([df_4, df_5, df_6])
```

```python
len(df_test)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_train.drop(['class'],axis=1))
predictions=kmeans.predict(df_test.drop(['class'],axis= 1))
```

```python
predictions
```

```python
score = accuracy_score(df_test['class'], predictions)
score
```

```python
print(classification_report(df_test['class'], predictions)) 
```

### Normalizing

```python
df_drop = df.drop(['class'],axis = 1)
df_drop
```

```python
mn=preprocessing.MinMaxScaler()
x_norm =mn.fit_transform(df_drop)
```

```python
x_norm
```

```python
dataset = pd.DataFrame({'sepalLength': x_norm[:, 0], 'sepalWidth': x_norm[:, 1], 'petalLength': x_norm[:, 2], 'petalWidth': x_norm[:, 3]})
```

```python
dataset
```

```python
dataset_1=dataset.iloc[0:35,]
dataset_4=dataset.iloc[35:50,]
dataset_2=dataset.iloc[50:85,]
dataset_5=dataset.iloc[85:100,]
dataset_3=dataset.iloc[100:135,]
dataset_6=dataset.iloc[135:150,]
```

```python
data_train=pd.concat([dataset_1, dataset_2, dataset_3])
data_test=pd.concat([dataset_4, dataset_5, dataset_6])
```

```python
data_test
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(data_train)
```

```python
predictions=kmeans.predict(data_test)
score = accuracy_score(df_test['class'], predictions)
score
```

```python
print(classification_report(df_test['class'], predictions)) 
```
